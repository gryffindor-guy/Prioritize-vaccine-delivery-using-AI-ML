```
# !pip install geopandas
from geopandas import GeoDataFrame, read_file
map_data = read_file('Indian_States.shp')
map_data.rename(columns = {"st_nm":"States/UT"}, inplace = True)
map_data.head()
```

```
fig, ax = plt.subplots(1, figsize=(20, 12))
ax.axis("off")

df = GeoDataFrame(df, geometry= map_data['geometry'] )
ax.set_title("Covid-19 Statewise Data — Confirmed Cases", 
             fontdict =  {"fontsize" : "25", "fontweight" : "3"})
df.plot(column = "Confirmed", cmap="YlOrRd", 
                 linewidth=0.8, ax=ax, edgecolor="0.8", legend = True)
plt.show()
```

```
sns.set_style("ticks")
plt.figure(figsize = (800,100))
plt.barh(df["District"],    df["Confirmed"].map(int),align = 'center', color = 'lightblue', edgecolor = 'blue')
plt.xlabel('No. of Confirmed cases', fontsize = 18)
plt.ylabel('States/UT', fontsize = 18)
plt.gca().invert_yaxis()
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.title('Total Confirmed Cases Statewise', fontsize = 18 )
for index, value in enumerate(df['Confirmed']):
    plt.text(value, index, str(value), fontsize = 12)
plt.show()
```
